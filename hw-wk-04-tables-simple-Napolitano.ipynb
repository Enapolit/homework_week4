````python
### Find all the registration data
registrations = soup.find_all("p", class_="registration")
````

````python
registrations
````

### Place all the registration data into a list with only the numbers in the format.
It should look like this:

```['4235', '4234', '4231']```

````python
## do it here (create more cells if you need them)
registration_data = []
for registration in registrations:
    registration.get_text()
    registration_data.append(registration.get_text())
````

````python
registration_data
````

````python
##registration_data #list comprehension 

##test = registration_data[0]
##test.strip("Registration# ")

registration_list = []
for number in registrations:
    number.strip("Registration# ")
    registration_list.append(number.strip("Registration# "))
````

````python
registration_list
````

### Place all the company names into a list.
It should look like this:

```['Recovery Foundation','New Horizons','Renewable Light']```

````python
## do it here (create more cells if you need them)
companies = soup.find_all("a")
companies
````

````python
c_names = []
for company in companies:
    company.get_text()
    c_names.append(company.get_text())
c_names
````

````python
### Do the same but via list comprehension
company_list = [company.get_text() for company in companies]
company_list
````

### Place all the status into a list.
It should look like this:

```['Active', 'Inactive', 'Active']```

````python
## do it here (create more cells if you need them) ##CORRECTION: 
##Place all the cost into a list.¶
##It should look like this:

##['$10,000', '$15,000', '$11,000']
costs = soup.find_all("p", class_="cost")
costs
````

````python
cost_list = [cost.get_text() for cost in costs]
cost_list
````

### Place all the inspection results into a list.
It should look like this:

```['Passed', 'Failed', 'Passed']```

````python
inspections = soup.find_all("p", class_="status")
inspections
````

````python
inspection_list = [inspection.get_text() for inspection in inspections]
inspection_list
````

````python
recovery_data = []
for (inspection, cost, companies, registration) in zip(inspection_list, cost_list, company_list, registration_data):
    recovery_data.append({
        "inspection": inspection_list,
        "cost": cost_list,
        "companies": company_list,
        "registration": registration_data,
    })
    
recovery_data

## Turn these 4 lists into a dataframe 
````

````python
df = pd.DataFrame(recovery_data)
````

````python
df
````

````python
## export the dataframe to a csv file. 
df.to_csv("recovery_centers.csv", encoding = "UTF-8", index = False)
````
